```python
import dlib as dlib
import matplotlib.pyplot as plt
import cv2
import random
```

### Step 1

Load an image (any image that contains faces).

```python
image = cv2.imread('/content/drive/MyDrive/people3.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap='gray' )
```

### Step 2

Load the dlib face predictor.

```python
face_detector = dlib.get_frontal_face_detector()
```

### Step 3

Run the predictor on your image.

```python
faces = face_detector(gray_image, 0)
```

```python
for bbox in faces:
    x1 = bbox.left()
    y1 = bbox.top()
    x2 = bbox.right()
    y2 = bbox.bottom()
    random_number1 = random.randint(0, 255)
    random_number2 = random.randint(0, 255)
    random_number3 = random.randint(0, 255)
    cv2.rectangle(image, (x1, y1), (x2, y2), (random_number1, random_number2, random_number3), 2)
```

```python
plt.imshow(image)
```
